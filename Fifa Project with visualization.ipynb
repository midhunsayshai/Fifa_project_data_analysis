### Clearing the data

### Dropping the first column named 'Unnamed'

```python
fifa.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
fifa
```

### To plot top 5 salaried players

```python
fifa['Wage'].dtype
```

```python
fifa['Wage']=fifa['Wage'].str.replace('€','')
fifa['Wage']=fifa['Wage'].str.replace('K','000')
```

```python
fifa
```

```python
fifa['Wage']=fifa['Wage'].astype(int)
fifa['Wage'].dtype
```

```python
Highpaid=fifa.sort_values(by='Wage', ascending=False)[['Name', 'Wage']].head(5)
```

```python
Highpaid
```

### Visualizing the Highpaid Dataframe

```python
xplot=np.array(["L.Messi","L.Suarez","L.Modric","C.Ronaldo","S.Ramos"])
yplot=np.array([565000,455000,420000,405000,380000])
mycolor=["Red","Black","Pink","Violet","Yellow"]
plt.title("Highly Paid")
plt.xlabel("Player Name")
plt.ylabel("Wage")
plt.bar(xplot,yplot,color=mycolor)
plt.show()
```

### To Plot lowest 5 Salaried Players

```python
fifa['Wage'].isnull().sum()
```

```python
Lowest=fifa.sort_values(by='Wage',ascending=False)[['Name', 'Wage']].tail(5)
Lowest
```

```python
fifa['Wage'].mean()
```

```python
fifa['Wage']=fifa['Wage'].replace(0,9732)
fifa['Wage']
```

```python
Low=fifa[fifa['Wage']>9732]
Low
```

```python
LowestSalaried = Low.sort_values(by='Wage', ascending=True)[['Name', 'Wage']].head(5)
LowestSalaried
```

### 2. To plot total salary

```python
fifa['Wage'].dtype
```

```python
Salarytrend=fifa['Wage']
Salarytrend
```

```python
plt.figure(figsize=(8,4))
sns.histplot(Salarytrend, bins=20)
plt.title("Salary Range of Players")
plt.xlabel("Wage")
plt.ylabel("Number of Players")
plt.show()
```

### 3. To plot the count of preferred foot.

```python
fifa.columns
```

```python
fifa['Preferred Foot'].value_counts
```

```python
Prefer=fifa['Preferred Foot'].value_counts()
Prefer
```

```python
Prefer.values
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=Prefer.index, y=Prefer.values)
plt.title("Preferred Foot for Players")
plt.xlabel("Preferred Foot")
plt.ylabel("Number of Players")
plt.show()
```
